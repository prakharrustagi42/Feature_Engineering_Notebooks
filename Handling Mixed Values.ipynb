```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("titanic_mixed_data.csv")
```

```python
df.head()
```

```python
## the entire goal is to split this cabin C85 into c and 85 and similarly 
## for the ticket number and also the number columns
```

```python
df.info()
```

```python
df["Cabin_coach"] = df["Cabin"].str.extract("(\d+)")

df["Cabin_coach"] = pd.to_numeric(df["Cabin_coach"], errors="coerce", downcast="integer")
```

```python
df["Cabin_number"] = df["Cabin"].str[0]
```

```python
df
```

```python
## now getting to the number column 

df["number_integer"]= pd.to_numeric(df["number"], errors = "coerce", downcast = "integer")
```

```python
df["number_letter"]= np.where(df["number_integer"].isnull(),df["number"],0)
```

```python
df
```

```python
## similarly we get to the ticket number 


df["Ticket_number"]= df["Ticket"].apply(lambda s: s.split()[-1])
df["Ticket_number"] = pd.to_numeric(df["Ticket_number"], errors = "coerce", downcast = "integer")
```

```python
df["Ticket"].isnull().sum() ## no zeros here 
```

```python
df["Ticket_class"]= df["Ticket"].apply(lambda s: s.split()[0])

## in the above some lists just have numbers so class will have ticket numbers . to correct that 

df["Ticket_class"] = np.where(df["Ticket_class"].str.isdigit(), np.nan, df["Ticket_class"])
## rhere str class is used because we are checking if a string is a number or not 
```

```python
df ## everything seems sorted 
```

```python
df.drop(columns = ["Cabin","Ticket","number"], inplace = True)
```
